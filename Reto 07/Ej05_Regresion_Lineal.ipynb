```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


from google.colab import files
uploaded = files.upload()


df = pd.read_excel('IBEX35_Sept2018.xls')


X = df[['Apertura']]
y = df['Cierre']

modelo = LinearRegression()
modelo.fit(X, y)
y_pred = modelo.predict(X)

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Ecuación: y = {modelo.intercept_:.2f} + {modelo.coef_[0]:.2f} * Apertura")
print(f"Error cuadrático medio: {mse:.2f}")
print(f"R²: {r2:.2f}")

plt.scatter(X, y, label="Valores reales")
plt.plot(X, y_pred, color="red", label="Recta ajustada")
plt.xlabel("Apertura")
plt.ylabel("Cierre")
plt.title("Regresión lineal cierre vs apertura bolsa")
plt.legend()
plt.show()
```
