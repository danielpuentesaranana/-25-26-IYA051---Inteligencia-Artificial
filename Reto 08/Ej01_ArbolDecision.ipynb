```python
# Árbol de decisión básico para clasificar animales (hasta 5 niveles de profundidad)
# Basado en las imágenes de vertebrados e invertebrados

def clasificar_animal():
    print("Responde con 'sí' o 'no'.")

    columna_vertebral = input("¿El animal tiene columna vertebral? ")
    if columna_vertebral.lower() == "sí":
        # Vertebrados: 2º nivel
        tipo_cuerpo = input("¿Tiene plumas, pelo, piel húmeda o escamas? (plumas/pelo/piel húmeda/escamas) ")
        if tipo_cuerpo.lower() == "pelo":
            print("→ Mamífero")
        elif tipo_cuerpo.lower() == "escamas":
            extremidad = input("¿Tiene aletas o patas? (aletas/patas) ")
            if extremidad.lower() == "aletas":
                print("→ Pez")
            else:
                print("→ Reptil")
        elif tipo_cuerpo.lower() == "plumas":
            print("→ Ave")
        elif tipo_cuerpo.lower() == "piel húmeda":
            print("→ Anfibio")
        else:
            print("No se puede clasificar.")
    else:
        # Invertebrados: 2º nivel
        proteccion = input("¿Tiene protección corporal? (sí/no) ")
        if proteccion.lower() == "sí":
            tipo_proteccion = input("¿Es una cubierta dura, concha, o espinas/placas? (cubierta/concha/espinas) ")
            if tipo_proteccion.lower() == "cubierta":
                patas = input("¿Cuántos pares de patas tiene? (3, 4, muchos, ninguno) ")
                if patas == "3":
                    print("→ Insecto (Artrópodo)")
                elif patas == "4":
                    print("→ Crustáceo (Artrópodo)")
                elif "muchos" in patas:
                    print("→ Miriápodo (Artrópodo)")
                else:
                    print("→ Arácnido (Artrópodo)")
            elif tipo_proteccion.lower() == "concha":
                concha = input("¿La concha es de dos piezas, no tiene, o es dorsal? (dos piezas/no tiene/dorsal) ")
                if concha == "dos piezas":
                    print("→ Bivalvo (Molusco)")
                elif concha == "dorsal":
                    print("→ Gasterópodo (Molusco)")
                else:
                    print("→ Cefalópodo (Molusco)")
            elif tipo_proteccion.lower() == "espinas":
                print("→ Equinodermo")
            else:
                print("No se puede clasificar.")
        else:
            # Sin protección corporal: 4º nivel
            cuerpo = input("¿El cuerpo está segmentado, es redondo, plano o con simetría irregular? (segmentado/redondo/plano/irregular) ")
            if cuerpo == "segmentado":
                print("→ Anélido (Gusano)")
            elif cuerpo == "redondo":
                print("→ Nematodo (Gusano)")
            elif cuerpo == "plano":
                print("→ Platelminto")
            elif cuerpo == "irregular":
                print("→ Porífero (Esponja)")
            else:
                tipo_celentereo = input("¿Es una medusa (flota en agua/formato campana) o pólipo (fijo a un objeto)? (medusa/pólipo) ")
                if tipo_celentereo == "medusa":
                    print("→ Medusa (Celentéreo)")
                else:
                    print("→ Pólipo (Celentéreo)")

clasificar_animal()
```
