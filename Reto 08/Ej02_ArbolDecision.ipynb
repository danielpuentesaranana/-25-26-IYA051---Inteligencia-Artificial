```python
def protocolo_covid_final():
    print("¿La persona tiene uno o más síntomas relacionados con COVID-19? (si/no)")
    sintomas = input().strip().lower()

    if sintomas == "si":
        print("¿Existe una condición crónica preexistente documentada que explique estos síntomas? (si/no)")
        cronica = input().strip().lower()
        if cronica == "si":
            print("✓ Puede regresar a la escuela/trabajo si la condición crónica explica completamente los síntomas Y está documentada.")
        else:
            print("¿La persona se ha realizado una prueba COVID? (si/no)")
            prueba = input().strip().lower()
            if prueba == "si":
                print("¿El resultado de la prueba molecular/antígeno, tomada después del inicio de los síntomas, fue negativo? (si/no)")
                negativo = input().strip().lower()
                if negativo == "si":
                    print("✓ Puede regresar cuando hayan pasado al menos 24 horas sin fiebre (sin medicamentos) y los síntomas hayan mejorado.")
                else:
                    print("¿El resultado fue positivo? (si/no realizada)")
                    resultado = input().strip().lower()
                    if resultado == "si" or resultado == "no realizada":
                        print("¿Puede y está dispuesta a usar mascarilla días 6-10? (si/no)")
                        mascarilla = input().strip().lower()
                        if mascarilla == "si":
                            print("✗ No puede regresar todavía. Deberá esperar hasta el día 6 desde el inicio de síntomas, siempre que esté sin fiebre 24h y hayan mejorado los síntomas. Además, debe usar mascarilla bien ajustada los días 6-10 tras volver.")
                        else:
                            print("✗ No puede regresar hasta el día 11 desde inicio de síntomas (sin fiebre 24h y síntomas mejorando).")
                    else:
                        print("✗ No puede regresar todavía (caso positivo).")
            else:
                print("✗ No puede regresar todavía. Debe quedarse en casa hasta que mejoren los síntomas y hayan pasado al menos 24 horas sin fiebre (sin medicación).")
    else:
        print("¿La persona fue identificada como contacto cercano de un positivo en los últimos 14 días? (si/no)")
        contacto = input().strip().lower()
        if contacto == "si":
            print("→ Seguir el protocolo para 'Contacto cercano y cuarentena'.")
        else:
            print("✓ Puede regresar a la escuela/trabajo normalmente.")

protocolo_covid_final()
```
