```python
from sklearn.datasets import load_iris
from sklearn import tree
import matplotlib.pyplot as plt

iris = load_iris()
clf = tree.DecisionTreeClassifier()
clf.fit(iris.data, iris.target)

# Visualizacin del rbol
plt.figure(figsize=(12,8))
tree.plot_tree(clf, feature_names=iris.feature_names,
               class_names=iris.target_names, filled=True, rounded=True)
plt.show()

# Predicciones para varios ejemplos, puedes cambiar estos valores
ejemplos = [
    [5.1, 3.5, 1.4, 0.2],  # Iris-setosa
    [6.7, 3.1, 4.4, 1.4],  # Iris-versicolor
    [6.3, 2.9, 5.6, 1.8],  # Iris-virginica
    [4, 2, 1, 0.2]         # Cambia a lo que quieras probar
]

for e in ejemplos:
    pred = clf.predict([e])
    print(f"Valores: {e} -> Prediccin: {iris.target_names[pred][0]}")
```
